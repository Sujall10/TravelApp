```python
def assign_rating(text):
    if any(word in text.lower() for word in ["excellent", "great", "luxurious", "smooth"]):
        return 5
    elif any(word in text.lower() for word in ["okay", "average", "decent"]):
        return 3
    elif any(word in text.lower() for word in ["terrible", "worst", "horrible", "dirty", "broke down"]):
        return 1
    elif any(word in text.lower() for word in['null']):
        return 0
    else:
        return 2 
```

```python
Review["flight_rating"] = Review["review_flights"].apply(assign_rating)
Review["hotel_rating"] = Review["review_hotel"].apply(assign_rating)
Review["car_rating"] = Review["review_car"].apply(assign_rating)
```

```python
Review
```

```python
def calculate_overall_rating(row):
    if row["car_rating"] == 0:
        return row[["flight_rating", "hotel_rating"]].mean()
    else:
        return row[["flight_rating", "hotel_rating", "car_rating"]].mean()
```

```python
Review['overall_rating'] = Review.apply(calculate_overall_rating, axis=1)
```
